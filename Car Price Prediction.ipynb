```python
# Split the Data into Training & Testing sets.

from sklearn.model_selection import train_test_split
car_data = car_price.drop(['car_ID','price','CarName','fueltype','aspiration','doornumber','carbody','drivewheel','enginelocation',
                          'enginetype','cylindernumber','fuelsystem'], axis=1)
car_data
```

```python
car_data = np.array(car_data)
car_data
```

```python
car_label = np.array(car_price['price'])
car_label
```

```python
x_train,x_test,y_train,y_test = train_test_split(car_data, car_label, test_size=0.2, random_state=10)
x_train
```

```python
y_train
```

```python
x_test
```

```python
y_test
```

```python
# Train the Model(Decision Tree Regressor)

from sklearn.tree import DecisionTreeRegressor
DTR = DecisionTreeRegressor()
DTR.fit(x_train,y_train)
```

```python
# Make Predictions
new_data = [[ 2.000e+00,  9.610e+01,  1.768e+02,  6.660e+01,  5.050e+01,
         2.460e+03,  1.320e+02,  3.460e+00,  3.900e+00,  8.700e+00,
         9.000e+01,  5.100e+03,  2.300e+01,  3.100e+01]]
predi = DTR.predict(new_data)
predi
```

```python
y_predict = DTR.predict(x_test)
DTR.score(x_test,y_predict)
```
